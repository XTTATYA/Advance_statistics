```python
#Q5)Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
"""Ans"""
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu0=0, mu1=1, sigma=1, alpha=0.05, test_type='right'):

    x = np.linspace(mu0 - 4*sigma, mu1 + 4*sigma, 1000)
    y0 = norm.pdf(x, mu0, sigma)
    y1 = norm.pdf(x, mu1, sigma)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y0, label='H₀ distribution', color='blue')
    plt.plot(x, y1, label='H₁ distribution', color='green')

    if test_type == 'right':
        critical_value = norm.ppf(1 - alpha, mu0, sigma)
        plt.fill_between(x, 0, y0, where=(x > critical_value), color='red', alpha=0.4, label='Type I error (α)')
        plt.fill_between(x, 0, y1, where=(x <= critical_value), color='orange', alpha=0.4, label='Type II error (β)')
        plt.axvline(critical_value, color='black', linestyle='--', label='Critical value')

    elif test_type == 'left':
        critical_value = norm.ppf(alpha, mu0, sigma)
        plt.fill_between(x, 0, y0, where=(x < critical_value), color='red', alpha=0.4, label='Type I error (α)')
        plt.fill_between(x, 0, y1, where=(x >= critical_value), color='orange', alpha=0.4, label='Type II error (β)')
        plt.axvline(critical_value, color='black', linestyle='--', label='Critical value')

    elif test_type == 'two':
        critical_low = norm.ppf(alpha / 2, mu0, sigma)
        critical_high = norm.ppf(1 - alpha / 2, mu0, sigma)
        plt.fill_between(x, 0, y0, where=(x < critical_low) | (x > critical_high), color='red', alpha=0.4, label='Type I error (α)')
        plt.fill_between(x, 0, y1, where=(x >= critical_low) & (x <= critical_high), color='orange', alpha=0.4, label='Type II error (β)')
        plt.axvline(critical_low, color='black', linestyle='--')
        plt.axvline(critical_high, color='black', linestyle='--')

    plt.title('Type I and Type II Errors in Hypothesis Testing')
    plt.xlabel('Test Statistic')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()
visualize_type1_type2_errors(mu0=0, mu1=1, sigma=1, alpha=0.05, test_type='right')
```

```python
#Q7) Perform a paired sample T-test using Python and visualize the comparison results
"""Ans"""
import numpy as np
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Example paired data (Before and After measurements)
before = np.array([85, 78, 90, 88, 76, 95, 89])
after = np.array([88, 82, 91, 90, 79, 97, 94])

# Calculate differences
diff = after - before

# Sample size and degrees of freedom
samplesize = len(diff)
df = samplesize - 1

# Hypotheses:
# H0: mean difference = 0
# H1: mean difference != 0

# Sample mean and std of differences
diff_mean = np.mean(diff)
diff_std = np.std(diff, ddof=1)

alpha = 0.05

# Calculate t-test statistic for paired sample
t_test = diff_mean / (diff_std / np.sqrt(samplesize))

# Calculate two-tailed critical t-value
t_crit = stats.t.ppf(1 - alpha/2, df)

# Decision making
if abs(t_test) < t_crit:
    print("We fail to reject the null hypothesis")
    print("Interpretation: No significant difference between paired samples")
else:
    print("We reject the null hypothesis")
    print("Interpretation: Significant difference between paired samples")


# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

sns.boxplot(data=[before, after])
plt.xticks([0,1], ['Before', 'After'])
plt.title('Paired Sample Comparison')

plt.ylabel('Values')
plt.show()
```

```python
#Q9)Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
"""Ans"""
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
def confidence_interval(sample, confidence=0.95):

    sample = np.array(sample)
    n = len(sample)
    mean = np.mean(sample)
    std_err = stats.sem(sample)#or np.std(sample,ddof=1)/np.sqrt(n)  # standard error = std / sqrt(n)
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_crit * std_err #formula for margin of error
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return (lower_bound, upper_bound)



# Example usage:
data = [102, 98, 100, 105, 97, 101, 99, 103]
ci = confidence_interval(data)
print(f"95% Confidence Interval for the mean: {ci}")
```

```python
#Q11) Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.
"""Ans"""
#consider an example :-A medical test for a disease is 99% accurate (true positive rate), but the disease only affects 1% of the population. If someone tests positive, what is the probability they actually have the disease?
#given
# Prior probability of having the disease
p_disease = 0.01
# Probability of testing positive given disease
p_pos_given_disease = 0.99
# Probability of testing positive given no disease
p_pos_given_no_disease = 0.05
# Prior probability of not having the disease
p_no_disease = 1 - p_disease

# Total probability of testing positive
p_positive = (p_pos_given_disease * p_disease) + (p_pos_given_no_disease * p_no_disease)

# Apply Bayes' Theorem
p_disease_given_positive = (p_pos_given_disease * p_disease) / p_positive

print(f"Probability of having disease given a positive test: {p_disease_given_positive:.4f}")
```

```python
#Q16)  Implement an F-test using Python to compare the variances of two random samples
"""Ans"""
import numpy as np
from scipy.stats import f

np.random.seed(42)
sample1 = np.random.normal(loc=10, scale=2, size=30)  # mean=10, std=2
sample2 = np.random.normal(loc=10, scale=3, size=30)  # mean=10, std=3

var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

if var1 > var2:
    F = var1 / var2
    dfn, dfd = len(sample1) - 1, len(sample2) - 1
else:
    F = var2 / var1
    dfn, dfd = len(sample2) - 1, len(sample1) - 1
alpha = 0.05
f_critical = f.ppf(1 - alpha/2, dfn, dfd)

if F > f_critical:
    print("Reject the null hypothesis — variances are significantly different.")
else:
    print("Fail to reject the null hypothesis — no significant difference in variances.")
```

```python
#Q17)Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results
"""Ans"""
import numpy as np
from scipy import stats

np.random.seed(42)
group1 = np.random.normal(75, 5, 30)  # mean=75, std=5, n=30
group2 = np.random.normal(80, 5, 30)  # mean=80, std=5, n=30
group3 = np.random.normal(78, 5, 30)  # mean=78, std=5, n=30

f_stat, p_value = stats.f_oneway(group1, group2, group3)

alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis — at least one group mean is different.")
else:
    print("Fail to reject the null hypothesis — no evidence of difference in means.")
```

```python
#Q18) Perform a one-way ANOVA test using Python to compare the means of different groups and plot the results.
"""Ans"""
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import f_oneway

np.random.seed(42)
group_A = np.random.normal(75, 5, 30)
group_B = np.random.normal(80, 5, 30)
group_C = np.random.normal(78, 5, 30)

data = pd.DataFrame({
    'Score': np.concatenate([group_A, group_B, group_C]),
    'Group': ['A'] * 30 + ['B'] * 30 + ['C'] * 30
})

f_stat, p_val = f_oneway(group_A, group_B, group_C)

if p_val < 0.05:
    print("Conclusion: Reject the null hypothesis — at least one group mean is significantly different.")
else:
    print("Conclusion: Fail to reject the null hypothesis — no significant difference in means.")

plt.figure(figsize=(8, 6))
sns.boxplot(x='Group', y='Score', data=data, palette="Set2")
plt.title("Group-wise Score Distribution")
plt.xlabel("Group")
plt.ylabel("Score")
plt.grid(True)
plt.show()
```

```python
#Q19)Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.
"""Ans"""
import numpy as np
import pandas as pd
from scipy.stats import shapiro, levene
import seaborn as sns
import matplotlib.pyplot as plt

def check_anova_assumptions(data, group_col, value_col):
    groups = data[group_col].unique()
    print("🔍 Normality Test (Shapiro-Wilk):")
    for group in groups:
        stat, p = shapiro(data[data[group_col] == group][value_col])
        print(f"  Group {group}: W = {stat:.4f}, p = {p:.4f} {'(Normal)' if p > 0.05 else '(Not Normal)'}")
    group_values = [data[data[group_col] == group][value_col] for group in groups]
    stat, p = levene(*group_values)
    print(f"\nEqual Variance Test (Levene): F = {stat:.4f}, p = {p:.4f} {'(Equal Variances)' if p > 0.05 else '(Unequal Variances)'}")

    print("\n Independence: Assumed based on proper experimental design or data collection.")

    sns.boxplot(x=group_col, y=value_col, data=data, palette="pastel")
    plt.title("Group-wise Boxplot")
    plt.show()

np.random.seed(0)
df = pd.DataFrame({
    'Score': np.concatenate([
        np.random.normal(70, 5, 30),
        np.random.normal(75, 5, 30),
        np.random.normal(72, 5, 30)
    ]),
    'Group': ['A'] * 30 + ['B'] * 30 + ['C'] * 30
})

check_anova_assumptions(df, 'Group', 'Score')
```

```python
#Q20)Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results
"""Ans"""
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols

np.random.seed(1)
n = 20
data = pd.DataFrame({
    'Score': np.random.normal(loc=75, scale=5, size=4 * n),
    'Method': np.repeat(['A', 'B'], repeats=2 * n),
    'Gender': np.tile(np.repeat(['Male', 'Female'], n), 2)
})

data.loc[(data['Method'] == 'B') & (data['Gender'] == 'Female'), 'Score'] += 5

model = ols('Score ~ C(Method) + C(Gender) + C(Method):C(Gender)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print("\n Two-Way ANOVA Table:\n")
print(anova_table)

plt.figure(figsize=(8, 6))
sns.boxplot(x='Method', y='Score', hue='Gender', data=data, palette="Set2")
plt.title("Interaction Effect of Method and Gender on Score")
plt.grid(True)
plt.show()
```

```python
#Q21)Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing
"""Ans"""
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

dfn = 5   # degrees of freedom numerator
dfd = 20  # degrees of freedom denominator

x = np.linspace(0.01, 5, 500)

y = f.pdf(x, dfn, dfd)

plt.figure(figsize=(8, 5))
plt.plot(x, y, 'b-', label=f'F-distribution (dfn={dfn}, dfd={dfd})')
plt.fill_between(x, y, alpha=0.2, color='blue')

alpha = 0.05
f_critical = f.ppf(1 - alpha, dfn, dfd)
plt.axvline(x=f_critical, color='red', linestyle='--', label=f'Critical value (α=0.05): {f_critical:.2f}')

plt.title("F-distribution Curve")
plt.xlabel("F value")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Q22) Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means.
"""Ans"""
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import f_oneway

np.random.seed(42)
group_A = np.random.normal(70, 5, 30)
group_B = np.random.normal(75, 5, 30)
group_C = np.random.normal(72, 5, 30)

df = pd.DataFrame({
    'Score': np.concatenate([group_A, group_B, group_C]),
    'Group': ['A'] * 30 + ['B'] * 30 + ['C'] * 30
})

f_stat, p_val = f_oneway(group_A, group_B, group_C)

print("🔍 One-Way ANOVA Result:")
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_val:.4f}")

if p_val < 0.05:
    print("Conclusion: Reject the null hypothesis — At least one group mean is different.")
else:
    print("Conclusion: Fail to reject the null hypothesis — No significant difference between group means.")

plt.figure(figsize=(8, 6))
sns.boxplot(x='Group', y='Score', data=df, palette='Set3')
plt.title("Group-wise Comparison of Scores")
plt.xlabel("Group")
plt.ylabel("Score")
plt.grid(True)
plt.show()
```

```python
#Q23) Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means
"""Ans"""
import numpy as np
from scipy import stats

np.random.seed(1)
population_mean = 100       # Known population mean (μ)
sample_size = 30
sample = np.random.normal(loc=102, scale=10, size=sample_size)  # Simulated sample with slight shift

sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)  # ddof=1 for sample standard deviation
alpha = 0.05
df = sample_size - 1

t_stat = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))
t_crit = stats.t.ppf(1 - alpha/2, df)

print("🔍 Hypothesis Testing (One-sample T-test)")
print(f"Sample Mean: {sample_mean:.2f}")
print(f"T-statistic: {t_stat:.4f}")
print(f"Critical T-value (α=0.05): ±{t_crit:.4f}")

if abs(t_stat) > t_crit:
    print("Conclusion: Reject the null hypothesis — sample mean is significantly different.")
else:
    print("Conclusion: Fail to reject the null hypothesis — no significant difference.")

import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(sample, kde=True, color='skyblue')
plt.axvline(population_mean, color='red', linestyle='--', label='Population Mean')
plt.axvline(sample_mean, color='green', linestyle='-', label='Sample Mean')
plt.title("Sample Distribution")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Q24) Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.
"""Ans"""
import numpy as np
from scipy.stats import chi2

np.random.seed(2)
sample = np.random.normal(loc=50, scale=8, size=25)  # mean=50, std dev=8
sample_size = len(sample)
sample_variance = np.var(sample, ddof=1)  # sample variance
assumed_variance = 64
chi2_stat = (sample_size - 1) * sample_variance / assumed_variance

alpha = 0.05
df = sample_size - 1
chi2_crit_low = chi2.ppf(alpha / 2, df)
chi2_crit_high = chi2.ppf(1 - alpha / 2, df)

print("🔍 Hypothesis Test for Population Variance")
print(f"Sample Variance: {sample_variance:.2f}")
print(f"Chi-square Statistic: {chi2_stat:.2f}")
print(f"Critical Region: χ² < {chi2_crit_low:.2f} or χ² > {chi2_crit_high:.2f}")

if chi2_stat < chi2_crit_low or chi2_stat > chi2_crit_high:
    print("Conclusion: Reject the null hypothesis — Population variance is significantly different.")
else:
    print("Conclusion: Fail to reject the null hypothesis — No significant difference in variance.")
```

```python
#Q25)Write a Python script to perform a Z-test for comparing proportions between two datasets or groups
"""Ans"""
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

successes = np.array([45, 30])     # e.g., 45 successes in group A, 30 in group B
samples = np.array([100, 100])     # e.g., 100 trials in both groups

z_stat, p_val = proportions_ztest(successes, samples)

print("🔍 Z-test for Comparing Two Proportions")
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_val:.4f}")

alpha = 0.05
if p_val < alpha:
    print("Conclusion: Reject the null hypothesis — Proportions are significantly different.")
else:
    print("Conclusion: Fail to reject the null hypothesis — No significant difference in proportions.")
```

```python
#Q26)Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results
"""Ans"""
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

np.random.seed(0)
group1 = np.random.normal(loc=50, scale=5, size=30)
group2 = np.random.normal(loc=50, scale=8, size=30)

var1 = np.var(group1, ddof=1)
var2 = np.var(group2, ddof=1)

f_stat = var1 / var2 if var1 > var2 else var2 / var1
dof1 = len(group1) - 1
dof2 = len(group2) - 1

alpha = 0.05
f_crit = f.ppf(1 - alpha/2, dof1, dof2)

print("🔍 F-Test for Comparing Variances")
print(f"Variance 1: {var1:.2f}")
print(f"Variance 2: {var2:.2f}")
print(f"F-statistic: {f_stat:.4f}")
print(f"Critical F-value (two-tailed, α = 0.05): {f_crit:.4f}")

if f_stat > f_crit:
    print("Conclusion: Reject the null hypothesis — Variances are significantly different.")
else:
    print("Conclusion: Fail to reject the null hypothesis — No significant difference in variances.")

plt.figure(figsize=(8, 5))
plt.hist(group1, bins=10, alpha=0.6, label='Group 1', color='blue')
plt.hist(group2, bins=10, alpha=0.6, label='Group 2', color='orange')
plt.title("Comparison of Group Distributions")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Q27) Perform a Chi-square test for goodness of fit with simulated data and analyze the results.
"""Ans"""
import numpy as np
from scipy.stats import chisquare
import matplotlib.pyplot as plt

observed = np.array([18, 17, 16, 19, 15, 15])  # simulated counts for each face of a die
total_observed = np.sum(observed)
expected = np.array([total_observed / 6] * 6) # expected if fair die (sum matches observed)

chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("🎲 Chi-square Goodness-of-Fit Test")
print(f"Observed Counts: {observed}")
print(f"Expected Counts: {expected}")
print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis — Data does NOT fit the expected distribution.")
else:
    print("Conclusion: Fail to reject the null hypothesis — Data fits the expected distribution.")

labels = ['1', '2', '3', '4', '5', '6']
x = np.arange(len(labels))
width = 0.35

plt.bar(x - width/2, observed, width, label='Observed', color='skyblue')
plt.bar(x + width/2, expected, width, label='Expected', color='orange')
plt.xticks(x, labels)
plt.xlabel("Dice Face")
plt.ylabel("Count")
plt.title("Observed vs Expected Counts (Goodness-of-Fit)")
plt.legend()
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
